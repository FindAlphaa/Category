```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# 주어진 데이터 프레임을 생성 (이 예시에서는 임의의 데이터로 생성)
df = df[df['Name'] != '삼성전자']
# df = df[df['Name'] != 'LG에너지솔루션']
# 스케일러 생성 및 적용
scaler = MinMaxScaler(feature_range=(300, 10000))
df['Marcap_scaled'] = scaler.fit_transform(df[['Marcap']])

print(df)
```

```python
# 'Name', 'Marcap', 'Marcap_scaled' 열만 선택하여 새로운 데이터프레임 생성
selected_df = df[['Name', 'Marcap', 'Marcap_scaled']]

# CSV 파일로 저장
selected_df.to_csv('시총_스케일러_re.csv', index=False)
```
